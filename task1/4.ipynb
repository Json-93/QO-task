```python
import numpy as np
from qutip import *

dim = 50
theta = np.pi / 4
amplitude = 2.5 * np.exp(1j * theta)
r = 0.5
xi = r * np.exp(1j * theta)

# ---- cat ---- 
ground = basis(dim, 0)
D_alpha = displace(dim, amplitude)
D_neg_alpha = displace(dim, -amplitude)

alpha_state = D_alpha * ground
neg_alpha_state = D_neg_alpha * ground
Cat_state_with_superposition = alpha_state - neg_alpha_state
Cat_state_with_superposition = Cat_state_with_superposition.unit()

# ---- kitten ---- 
squeezed_vacuum_state = squeeze(dim, xi) * ground

a_dagger = create(dim)
kitten_state = (a_dagger * squeezed_vacuum_state).unit()


# ---- product ----
sigma = ket2dm(Cat_state_with_superposition)
rho = ket2dm(kitten_state)

Fidelity_trace = fidelity(sigma, rho)**2
print(Fidelity_trace)
```
