```python
import numpy as np
import matplotlib.pyplot as plt
from qutip import *

# global parameter
dim = 25
amplitude = 2

# state
ground = basis(dim, 0)
D_alpha = displace(dim, amplitude)
D_neg_alpha = displace(dim, -amplitude)

alpha_state = D_alpha * ground
neg_alpha_state = D_neg_alpha * ground

# 1st question
Cat_state_without_superposition = 0.5 * ket2dm(alpha_state) + 0.5 * ket2dm(neg_alpha_state)


# plot
xvec = np.linspace(-5, 5, 150)
X, Y = np.meshgrid(xvec, xvec)

W = wigner(Cat_state_without_superposition, xvec, xvec)


fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, W, cmap='Reds', linewidth=0, antialiased=True)

# Colorbar
cbar = fig.colorbar(surf, ax=ax, shrink=0.5, aspect=15, pad=0.1)

ax.view_init(elev=30, azim=-45)
ax.set_title('Figure 1: Wigner Function (Without Superposition)')
ax.set_xlabel('Position (x)')
ax.set_ylabel('Momentum (p)')

plt.show()
```
