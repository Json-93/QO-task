```python
import numpy as np
import matplotlib.pyplot as plt
from qutip import *

# global parameter
dim = 25
theta = np.pi / 4
amplitude = 2.5 * np.exp(1j * theta)

# state
ground = basis(dim, 0)
D_alpha = displace(dim, amplitude)
D_neg_alpha = displace(dim, -amplitude)

alpha_state = D_alpha * ground
neg_alpha_state = D_neg_alpha * ground

# 2nd question
Cat_state_with_superposition = alpha_state - neg_alpha_state
Cat_state_with_superposition = Cat_state_with_superposition.unit() # 歸一化


xvec = np.linspace(-5, 5, 150)
X, Y = np.meshgrid(xvec, xvec)

W = wigner(Cat_state_with_superposition, xvec, xvec)

fig, ax = plt.subplots(figsize=(7, 6))
cont = ax.contourf(X, Y, W, levels=100)

cbar = fig.colorbar(cont, ax=ax)
cbar.set_label('Pseudoprobability')

ax.set_title('Figure 2: An optical Cat state with superposition')
ax.set_xlabel('Position (x)')
ax.set_ylabel('Momentum (p)')
ax.set_aspect('equal')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from qutip import *

# global parameter
dim = 25
theta = np.pi / 4
amplitude = 2.5 * np.exp(1j * theta)

# state
ground = basis(dim, 0)
D_alpha = displace(dim, amplitude)
D_neg_alpha = displace(dim, -amplitude)

alpha_state = D_alpha * ground
neg_alpha_state = D_neg_alpha * ground

# 2nd question
Cat_state_with_superposition = alpha_state - neg_alpha_state
Cat_state_with_superposition = Cat_state_with_superposition.unit() # 歸一化


# plot
xvec = np.linspace(-5, 5, 150)
X, Y = np.meshgrid(xvec, xvec)

W = wigner(Cat_state_with_superposition, xvec, xvec)


fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, W, cmap='bwr', linewidth=0, antialiased=True)

# Colorbar
cbar = fig.colorbar(surf, ax=ax, shrink=0.5, aspect=15, pad=0.1)

ax.view_init(elev=30, azim=-45)
ax.set_title('Figure 2: An optical Cat state with superposition 3D version')
ax.set_xlabel('Position (x)')
ax.set_ylabel('Momentum (p)')


plt.show()
```
